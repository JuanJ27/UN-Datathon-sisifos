```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
T1 = pd.read_csv("../../Data/VIVIENDA I.csv", sep="|")
T2 = pd.read_csv("../../Data/VIVIENDA II.csv", sep="|")
T3 = pd.read_csv("../../Data/VIVIENDA III.csv", sep="|")
T4 = pd.read_csv("../../Data/VIVIENDA IV.csv", sep="|")

T1 = T1[T1["area"] == 5]
T2 = T2[T2["area"] == 5]
T3 = T3[T3["area"] == 5]
T4 = T4[T4["area"] == 5]
```

```python
Ts = [T1, T2, T3, T4]
for T in Ts:
    T.rename(columns={"P8520S5":"acueducto",
                      "P8520S4":"basuras",
                      "P8520S3":"alcantarillado",
                      "P8520S2":"gas",
                      "P8520S1":"electricidad",
                      "P8520S1A1":"estrato"}, inplace=True)
```

```python
print(len(T1), len(T2), len(T3), len(T4))  
```

```python
T1S = T1.groupby("estrato",as_index=False)["estrato"].count()
T2S = T2.groupby("estrato",as_index=False)["estrato"].count()
T3S  = T3.groupby("estrato",as_index=False)["estrato"].count()
T4S = T4.groupby("estrato",as_index=False)["estrato"].count()  

T1S = T1S.iloc[1:]
T2S = T2S.iloc[1:]
T3S = T3S.iloc[1:]
T4S = T4S.iloc[1:]
```

```python
plt.plot(T1S.index,T1S["estrato"]/len(T1), label="T1")
plt.plot(T2S.index,T2S["estrato"]/len(T2), label="T2")
plt.plot(T3S.index,T3S["estrato"]/len(T3), label="T3")
plt.plot(T4S.index,T4S["estrato"]/len(T4), label="T4")
plt.legend()
plt.show()
```
