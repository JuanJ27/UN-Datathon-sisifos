```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../../Data/indice_pobreza_multidimen.csv", sep=",", encoding="latin-1")
```

```python
years = [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]
for i, row in df.iterrows():
    value_years = row[["i_2014", "i_2015", "i_2016", "i_2017", "i_2018", "i_2019", "i_2020", "i_2021", "i_2022", "i_2023"]].to_list()
    plt.plot(years, value_years, label=row["nombre"])
plt.show()
```
