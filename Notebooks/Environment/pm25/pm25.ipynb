### PM 25

#### Aranjuez

```python
AranjuezPM25 = pd.read_csv("../../../Data/aranjuez,-medell√≠n-air-quality.csv")
```

```python
AranjuezPM25["date"] = pd.to_datetime(AranjuezPM25["date"])
AranjuezPM25.sort_values(by="date", inplace=True)
AranjuezPM25Monthly = AranjuezPM25.resample("ME", on="date").mean().reset_index()
AranjuezPM25Yearly = AranjuezPM25.resample("YE", on="date").mean().reset_index()
```

```python
fig, ax = plt.subplots(3,1)
ax[0].plot(AranjuezPM25['date'], AranjuezPM25[' pm25'])
ax[0].set_title('Daily PM25')
ax[1].plot(AranjuezPM25Monthly['date'], AranjuezPM25Monthly[' pm25'])
ax[1].set_title('Monthly PM25')
ax[2].plot(AranjuezPM25Yearly['date'], AranjuezPM25Yearly[' pm25'])
ax[2].set_title('Yearly PM25')
plt.tight_layout()
```

```python
# Merge all the dataframes on the "date" column
all_places_df = AranjuezPM25[['date', ' pm25']].rename(columns={' pm25': 'AranjuezPM25'}).merge(
    PobladoPM25[['date', ' pm25']].rename(columns={' pm25': 'PobladoPM25'}),
    on='date', how='outer').merge(
    sanCristobalPM25[['date', ' pm25']].rename(columns={' pm25': 'sanCristobalPM25'}),
    on='date', how='outer').merge(
    villaHermosaPM25[['date', ' pm25']].rename(columns={' pm25': 'villaHermosaPM25'}),
    on='date', how='outer').merge(
    laCandelariaPM25[['date', ' pm25']].rename(columns={' pm25': 'laCandelariaPM25'}),
    on='date', how='outer').merge(
    belenPM25[['date', ' pm25']].rename(columns={' pm25': 'belenPM25'}),
    on='date', how='outer')
    

# Sort the dataframe by date
all_places_df.sort_values(by='date', inplace=True)

# Reset the index
all_places_df.reset_index(drop=True, inplace=True)
```

```python
all_places= all_places_df[(all_places_df['date'] >= '2020-01-01')]
```

```python
all_places.to_csv("../../../Data/datosPM25.csv")
```

```python
all_places_df = all_places_df.resample("ME", on="date").mean().reset_index()
```

```python
# Melt the dataframe to have 'place', 'date', and 'pm25' columns
melted_df = all_places_df.melt(id_vars=['date'], var_name='place', value_name='pm25')
# Drop rows with NaN values in 'pm25'
melted_df.dropna(subset=['pm25'], inplace=True)
# Reset the index
melted_df.reset_index(drop=True, inplace=True)
```
